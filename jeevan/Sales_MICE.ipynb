```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
sales_data = pd.read_csv("Retail_data.csv")
```

```python
sales_data.info()
```

```python
sales_data.isnull().sum()
```

```python
df = sales_data.apply(lambda series: pd.Series(
        LabelEncoder().fit_transform(series[series.notnull()]),
        index=series[series.notnull()].index))
```

```python
import time
start_time = time.time()

linear = LinearRegression()
imputer = IterativeImputer(estimator = linear, verbose = 2, max_iter = 50, tol =1e-10, imputation_order = 'roman')
imputer.fit(df)
transform = imputer.transform(df)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
sales_imputed_IterativeImputer = pd.DataFrame(transform, columns=df.columns)
sales_imputed_IterativeImputer.head()
```
