### detect the percentage of missing values in every column of the dataset #will give an idea about the distribution of missing values.


```python
mis_val =df.isna().sum()
mis_val_per = df.isna().sum()/len(df)*100
mis_val_table = pd.concat([mis_val, mis_val_per], axis=1)
mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
mis_val_table_ren_columns = mis_val_table_ren_columns[
       mis_val_table_ren_columns.iloc[:,:] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
mis_val_table_ren_columns
```

#### Experiment-1
#### KNN imputer for Null values

```python
import time
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer


start_time = time.time()

df = df.apply(lambda series: pd.Series(
        LabelEncoder().fit_transform(series[series.notnull()]),
        index=series[series.notnull()].index
    ))
 
imputer = KNNImputer(n_neighbors=3)
imputed =imputer.fit_transform(df)
```

```python
df_imputed_KNN = pd.DataFrame(imputed, columns=df.columns)
df_imputed_KNN
```

```python

Time_KNN = ("Time_KNN","--- %s seconds ---" % (time.time() - start_time))
Time_KNN
```

```python
df_imputed_KNN.isnull().sum()
```

#### Experiment-2
#### Statistical Imputation with the SimpleImputer with strategy Mean

```python
from sklearn.impute import SimpleImputer
start_time = time.time()
imputer = SimpleImputer(strategy = 'mean')
imputer.fit(df)
x_trans = imputer.transform(df)
x_trans
```

```python
df_imputed_SI_mean = pd.DataFrame(x_trans, columns=df.columns)
df_imputed_SI_mean
```

```python
Time_SI_mean= ("Time_SI_mean","--- %s seconds ---" % (time.time() - start_time))
Time_SI_mean
```

```python
df_imputed_SI_mean.isnull().sum()
```

#### Experiment-3
#### Statistical Imputation with the SimpleImputer with strategy Median

```python
start_time = time.time()
imputer = SimpleImputer(strategy = 'median')
imputer.fit(df)
x_trans = imputer.transform(df)
df_imputed_ST_median = pd.DataFrame(x_trans, columns=df.columns)
df_imputed_ST_median
```

```python
Time_SI_median = ("Time_SI_median","--- %s seconds ---" % (time.time() - start_time))
Time_SI_median
```

```python
df_imputed_ST_median.isnull().sum()
```

#### Experiment- 4
#### Statistical Imputation with the SimpleImputer with Most Frequent

```python
start_time = time.time()
imputer = SimpleImputer(strategy = 'most_frequent')
imputer.fit(df)
x_trans = imputer.transform(df)
df_imputed_ST_most_frequent = pd.DataFrame(x_trans, columns=df.columns)
df_imputed_ST_most_frequent
```

```python
Time_SI_most_frequent = ("Time_SI_most_frequent","--- %s seconds ---" % (time.time() - start_time))
Time_SI_most_frequent
```

```python
df_imputed_ST_most_frequent.isnull().sum()
```

#### Experiment- 5
#### MICE algorithm using Iterative Imputer to handle missing values

```python
from sklearn.linear_model import LinearRegression
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
start_time = time.time()
lr = LinearRegression()
imputer = IterativeImputer(estimator = lr, verbose = 2, max_iter = 50, tol =1e-10, imputation_order = 'roman')
imputer.fit(df)
x_trans = imputer.transform(df)
df_imputed_IterativeImputer = pd.DataFrame(x_trans, columns=df.columns)
df_imputed_IterativeImputer
```

```python
Time_MICE_IterativeImputer = ("Time_MICE_IterativeImputer", "--- %s seconds ---" % (time.time() - start_time))
Time_MICE_IterativeImputer
```
