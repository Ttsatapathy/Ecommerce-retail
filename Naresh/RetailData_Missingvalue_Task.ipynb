KNN IMPUTER

```python
from sklearn.impute import KNNImputer
from  sklearn.preprocessing import LabelEncoder
import time
```

```python
df1 = df.apply(lambda series: pd.Series(
        LabelEncoder().fit_transform(series[series.notnull()]),
        index=series[series.notnull()].index))
```

```python
imputer = KNNImputer(n_neighbors=3)
imputed_col=imputer.fit_transform(df1)
```

```python
start_time = time.time()
print("--- %s sec ---" % (time.time() - start_time))
```

```python
df_imputed_KNN = pd.DataFrame(imputed_col, columns=df1.columns)
df_imputed_KNN.head()
```

```python
df_imputed_KNN.isnull().sum()
```

**MICE** **IMPUTER**

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
start_time = time.time()
```

```python
lr = LinearRegression()
```

```python
lr = LinearRegression()
imputer = IterativeImputer(estimator = lr, verbose = 2, max_iter = 50, tol =1e-10, imputation_order = 'roman')
imputer.fit(df1)
```

```python
imputer.fit_transform(df1.drop([ 'order_name',
                          'product_type', 'product_title',
                          'product_vendor', 'variant_title',
                          'variant_title','api_client_title'] ,axis=1))
print("--- %s sec ---" % (time.time() - start_time))
```
